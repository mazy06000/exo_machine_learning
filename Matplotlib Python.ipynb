```python
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

# 1) Pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y)
```

```python
plt.figure()
plt.plot(x,y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure.png')
```

```python
plt.scatter(x,y) #nuage de point
```

# Subplot

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y, c='r')
plt.title('graphique 1')

plt.subplot(2,2,2)
plt.plot(x,np.sin(x),c='b')
plt.plot(x,np.cos(x),c='y')
plt.title('graphique2')
```

# OOP (non conseill)

```python
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
#opp plus interessante car plus de possibilit
fig, ax = plt.subplots(2,1,sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```

```python
dataset = {f'experience{i}': np.random.randn(100) for i in range(4)}
dataset
```

```python
def graphique(dataset):
    for i in range(len(dataset)):
        fig = dataset['experience'+str(i)]
        plt.figure()
        plt.subplot(4,1,i+1)
        plt.plot(fig)
```

```python
graphique(dataset)
```
